```python
import pandas as pd
```

```python
with open("../original_data/weather.txt") as file:
    data = file.read()
    
split_data = data.split('\n')
# print(split_data)
data_year_2010 = []
month = ""
number_of_months = 0

#w poniższej pętli tworzę dane 5 miesięcy z 2010 roku. Nie są one wyczyszczone.
for row in split_data:
    if row[11:15] == "2010" :
        if month != row[15:17]:
            number_of_months += 1
        if number_of_months <= 5:
            data_year_2010.append(row)
            month = row[15:17]
```

```python
#Następnie dane wiersze dzielę wg zmiennej na kolumny zgodnie z punktem 2.3 zawartym w pliku readme
#tworzę listy które w przyszłości będą kolumnami nowego df
ID_station_lst = []
year_lst = []
month_lst = []
examined_data = []
sign_lst = []
value_lst = []
flague_lst = []
day_lst = []
    
for row in data_year_2010:


    n = 1  #to jest numer dnia 
    i = 21 # i to zmienna która jest iteratorem po wierszu.
    
    while i != len(row):
        ID_station_lst.append(row[0:11])
        year_lst.append(row[11:15])
        month_lst.append(row[15:17])
        examined_data.append(row[17:21])
        sign_lst.append(row[i])
        i += 1
        value_lst.append(row[(i): (i+4)])
        i += 4
        flague_lst.append(row[(i): (i+3)])
        i += 3
        
        day_lst.append(n)
        n+= 1
```

```python
# Została stworzona wyczyszczona baza danych która zawiera następujące kolumny oraz spełnia warunki:
# Każda zmienna tworzy kolumnę.
# Każda obserwacja tworzy rząd.

tidied_data = {"ID_station" : ID_station_lst, "year" : year_lst, "month" : month_lst, "day" : day_lst, "measured_datatype": examined_data,
              "sign" : sign_lst , "value": value_lst, "flague": flague_lst, "day" : day_lst}

df = pd.DataFrame(tidied_data)
df
```

```python
# Aby spełnić trzeci warunek: Każdy typ jednostki obserwacyjnej tworzy tabelę. Tworzę DataFrame bez kolumny ID_station i year
days_weather_measur_2010 = df[["day", "month", "measured_datatype", "value", "flague"]]
days_weather_measur_2010
```

```python
#zapisuje do folderu
days_weather_measur_2010.to_csv(path_or_buf="../analisys_data/tab1")
```

```python
#następnie usuwam wiersze bez pomiary tzn w których wartość zmierzona jest równa 9999
days_weather_measured_2010 = days_weather_measur_2010[days_weather_measur_2010["value"] != "9999"]
days_weather_measured_2010
```
