Ćwiczenie 3:
ćwiczenie zawiera funkcję plotowaną na jednym wykresie z legendą. Na następnych wykresach widać plotowanie w osobnych figurach.

```python
def fun(x):
    return x**2+5

x1 = np.linspace(-1, 1)
x2 = np.linspace(-6, 6)
x3 = np.linspace(0, 5)
y = fun(x1)
plt.figure()
plt.plot(x1, fun(x1), label="y1")
plt.plot(x2, fun(x2), label="y2")
plt.plot(x3, fun(x3), label="y3")

plt.ylabel("oś y")
plt.xlabel("oś x")
plt.title("wykres 1")
plt.legend()
```
